```python
import pandas as pd
import numpy as np
from pandas import DataFrame, Series
```

```python
df_players = pd.read_csv('Players.csv')
```

```python
df_players['Player']
```

```python
len(df_players)
```

```python
names = []
for i in range(len(df_players)):
    names.append(df_players.iloc[i]['Player'])
names
```

```python
cleanedList = [x for x in names if str(x) != 'nan']
cleanedList
```

```python
cleanedList[1]
```

```python
for n in range(len(names) - 1):
    cleanedList[n] = cleanedList[n].replace('*','')
```

```python
firsts = []
lasts = []
for string in cleanedList:
    name = string.split(" ")
    fname = name[0]
    lname = name[1]
    firsts.append(fname)
    lasts.append(lname)
```

```python
files = open('fnames2.txt','w')
```

```python
for name in firsts:
    files.write('"')
    files.write(name)
    files.write('"')
    files.write(',') 
```

```python
files = open('lnames2.txt','w') 
```

```python
for name in lasts:
    files.write('"')
    files.write(name)
    files.write('"')
    files.write(',') 
```
